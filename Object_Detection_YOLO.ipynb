## **Import Libraries**

```python
import cv2
import supervision as sv

from tqdm import tqdm
from inference.models.yolo_world.yolo_world import YOLOWorld

from gtts import gTTS
from IPython.display import Audio, display, Video
import tempfile
from tqdm import tqdm
```

## **Example Data**

```python
# SOURCE_IMAGE_PATH = f"{HOME}/drive/MyDrive/Visio Sense Project/dog.jpeg"
# SOURCE_VIDEO_PATH = f"{HOME}/drive/MyDrive/Visio Sense Project/yellow-filling.mp4"

# SOURCE_IMAGE_PATH = f"{HOME}/drive/MyDrive/Visio Sense Project/dog1.jpg"
SOURCE_VIDEO_PATH = f"{HOME}/drive/MyDrive/Visio Sense Project/room.mp4"
```

## **Define The Model**

```python
model = YOLOWorld(model_id="yolo_world/l")
```

```python
TARGET_VIDEO_PATH = f"{HOME}/drive/MyDrive/Visio Sense Project/room_new.mp4"

frame_generator = sv.get_video_frames_generator(SOURCE_VIDEO_PATH)
video_info = sv.VideoInfo.from_video_path(SOURCE_VIDEO_PATH)

width, height = video_info.resolution_wh
frame_area = width * height
```

```python
# Set the class names for the model
classes = ["vehicle", "fog", "person", "table", "chair", "door", "flower vaas"]
model.set_classes(classes)
```

```python
# Temporary directory for audio files
temp_dir = tempfile.mkdtemp()
```

Repeat objects

```python
from gtts import gTTS
import os
import time
from collections import Counter
from IPython.display import Audio, display

BOUNDING_BOX_ANNOTATOR = sv.BoundingBoxAnnotator(thickness=2)
LABEL_ANNOTATOR = sv.LabelAnnotator(text_thickness=2, text_scale=1, text_color=sv.Color.BLACK)

# Processing video and saving annotated version
frame_rate = video_info.fps  # Get the frame rate of the video
frame_interval = frame_rate  # Number of frames to skip to process 1 frame per second

with sv.VideoSink(target_path=TARGET_VIDEO_PATH, video_info=video_info) as sink:
    for i, frame in enumerate(tqdm(frame_generator, total=video_info.total_frames)):
        # Skip frames to achieve 1 frame per second processing
        if i % frame_interval != 0:
            continue

        # Run inference on the selected frame
        results = model.infer(frame, confidence=0.002)
        detections = sv.Detections.from_inference(results).with_nms(threshold=0.1)
        detections = detections[(detections.area / frame_area) < 0.10]

        # Annotate the frame with detections
        annotated_frame = frame.copy()
        annotated_frame = BOUNDING_BOX_ANNOTATOR.annotate(annotated_frame, detections)
        annotated_frame = LABEL_ANNOTATOR.annotate(annotated_frame, detections)
        sink.write_frame(annotated_frame)

        # Count the number of each detected class
        class_counts = Counter()
        for detection in detections:
            class_id = detection[5]  # Assuming class ID is in the 5th position
            label = class_id['class_name']
            class_counts[label] += 1

        # Provide feedback for each detected class
        for label, count in class_counts.items():
            if label in classes:  # Filtering only for specific classes like "vehicle"
                if count == 1:
                    feedback_text = f"There is a {label} in front of you."
                elif count == 2:
                    feedback_text = f"There are two {label}s in front of you."
                else:
                    feedback_text = f"There are many {label}s in front of you."

                # Generate voice feedback
                tts = gTTS(feedback_text)

                # Save audio file to temporary directory
                audio_file_path = os.path.join(temp_dir, f"{label}.mp3")
                tts.save(audio_file_path)

                # Play the audio
                display(Audio(audio_file_path, autoplay=True))

                # Add a delay for clearer feedback
                time.sleep(4)  # Adjust the time (in seconds) as needed

# After the processing is complete, display the saved video
if os.path.exists(TARGET_VIDEO_PATH):
    print("Video saved successfully at:", TARGET_VIDEO_PATH)
    display(Video(TARGET_VIDEO_PATH, embed=True))
else:
    print("Error: Video not found at the specified path.")
```
